```python
import pandas as pd
import numpy as np
```

```python
from sklearn import datasets
datasets.__all__
```

```python
data = datasets.load_iris()
df = pd.DataFrame(data.data)
df.columns = data.feature_names
df
```

```python
df['target'] = data.target
df
```

```python
df0 = df[df.target==0]
df0
```

```python
df1 = df[df.target==1]
df1
```

```python
df2 = df[df.target==2]
df2
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Sepal Lenght (in cm)')
plt.ylabel('Sepal Length (in cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],s=5,c='green')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],s=5,c='red')
```

```python
plt.xlabel('petal Lenght (in cm)')
plt.ylabel('Petal Length (in cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],s=5,c='green')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],s=5,c='red')
```

```python
X = df.drop(['target'],axis=1)
y = df.target
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=24)
```

```python
len(X_train)
```

```python
len(X_test)
```

Training model

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf')
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
